# Useless Animals Sorting

A bored zookeeper, tired of the monotony of his daily routine, decided to invent a machine to sort the animals arriving at the zoo. Not that it would make his life any easier... he's just bored out of his mind and needs something to pass the time. An amateur engineer in his spare time, he has begun designing the contraption and now turns to you for help. You look just as bored as he is and everyone loves machines right ? So why not lend a hand?



## Observations:
We see that the bot returns us the solution on error and we can have up to 5 errors per connection.

## Approach:
We can just send any animal and store the response by matching with image hash.

```python
import pwn, time, hashlib, base64, os


def get_image(img):
    b64 = base64.b64decode(img)
    
    hash = hashlib.md5(b64).hexdigest()
    
    with open(f'images/default.jpg', 'wb') as f:
        f.write(b64)
    
    return hash

def get_solution(hash):
    if not os.path.exists(f'solutions/{hash}.txt'):
        return "dog"
    with open(f'solutions/{hash}.txt', 'r') as f:
        return f.read()

def process(stats):
    # open
    p = pwn.remote('challenges.hackday.fr', 51259)

    (p.recvuntil(b'the game'))
    p.sendline(b'')
    (p.recvuntil(b': \n'))



    while True:
        image = p.recvline()
        p.recvline()
        img = get_image(image)
        solution = get_solution(img)
        stats['n_total'] += 1
        
        p.sendline(solution)
        sol = (p.recvline())
        if b'Correct !' in sol:
            print(sol)
            stats['n_correct'] += 1
        elif b'solution is' in sol:
            print(sol)
            correct_sol = sol.split(b' ')[5].decode()
            print(f'Correct solution is {[correct_sol]}')
            with open(f'solutions/{img}.txt', 'w') as f:
                f.write(correct_sol)
        else:
            print(sol)
        print(p.recvline())

while True:
    d = {'n_total': 0, 'n_correct': 0}
    try:
        process(d)
    except:
        pass
    print(d)
    time.sleep(1)
    
```

From what I've seen there were approx. 1k images, so it was fast to get their solution.
